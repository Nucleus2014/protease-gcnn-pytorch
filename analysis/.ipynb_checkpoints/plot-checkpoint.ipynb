```python
def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    try:
        x = x.numpy().T
    except AttributeError:
        x = x.T
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)
```

```python
def accuracy_for_each_ternary_class(f):
    name = '_'.join(f.split("_")[2:4])
    #logits = pkl.load(open(f,'rb'))
    logits = np.loadtxt(f)
    logits = softmax(logits).T
    labels = np.argmax(logits, axis=1)
    true_label = pkl.load(open("/Users/cplu/Downloads/Documents/RESEARCH/ml-cleavage/Data/y_test_" + name + "_0.3","rb"))
    classes = []
    for i in true_label.values:
        if i == 'CLEAVED':
            classes.append(0)
        elif i == 'UNCLEAVED':
            classes.append(1)
        elif i == 'MIDDLE':
            classes.append(2)
    true_label = np.asarray(classes)
    print(true_label.shape[0])
    print(labels.shape[0])
    assert true_label.shape[0] == labels.shape[0]
    print('total accuracy:{}'.format(np.mean(true_label == labels)))
    cleaved_mask = np.asarray(true_label == 0, dtype=bool)
    middle_mask = np.asarray(true_label == 2, dtype=bool)
    uncleaved_mask = np.asarray(true_label == 1, dtype=bool)
    print("number of samples in cleaved, middle, uncleaved classes")
    print(np.sum(cleaved_mask),np.sum(middle_mask),np.sum(uncleaved_mask))
    return np.mean(true_label[cleaved_mask] == labels[cleaved_mask]),np.mean(true_label[middle_mask] == labels[middle_mask]),np.mean(true_label[uncleaved_mask] == labels[uncleaved_mask])
```

# Variable Importance

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
os.chdir('/Users/cplu/Downloads/Documents/RESEARCH/GCNN/protease-gcnn-pytorch/outputs')
def importance_plot(filename, mode='accuracy',plot='all'):
    fp = open(os.path.join('variable_importance',filename),"r")
    accs = []
    original_acc = 0
    num_node = 0
    for line in fp:
        if line[0:8] == "Original":
            original_acc = float(line[line.index("(") + 1:].split(",")[0])
        elif line[0:4] == "Node" or line[0:4] == "Edge":
            accs.append(float(line.strip().split(" ")[-1]))
            if line[0:4] == "Node":
                num_node += 1
    assert mode == 'error' or 'accuracy'
    if mode == 'accuracy':
        tmp = [(original_acc - x) / original_acc for x in accs]
        normalized = [y/np.sum(tmp) for y in tmp]
    elif mode == 'error':
        tmp = [(x - original_acc)/ (1 - original_acc) for x in accs]
        normalized = [y/np.sum(tmp) for y in tmp]
    edge_ind = []
    for ind in range(1,num_node+1):
        k = ind + 1
        while k < num_node+1:
            edge_ind.append((ind,k))
            k += 1

#     nodes = [198,199,200,201,202,58,70,72,73,96,138,147,150,151,152,154,170,171,172,173,174,175,176,177,178,179,180,181,182,183,197,203,204,205]
    nodes = ['P6','P5','P4','P3','P2',58,70,72,73,96,138,147,150,151,152,154,170,171,172,173,174,175,176,177,178,179,180,181,182,183,"P7","P1","P1'","P2'"]
    df = pd.DataFrame(normalized, index=range(len(normalized)))
    new_index = []
    for i in range(1,num_node+1):
        new_index.append(str(nodes[i-1]))
#         new_index.append("Node" + str(i))
    for j in range(1,len(accs) - num_node + 1):
        pair = edge_ind[j-1]
        new_index.append(' '.join([str(nodes[pair[0]-1]), str(nodes[pair[1]-1])]))
#         new_index.append("Edge" + str(j))
    df.index = new_index
    df_sort = df.sort_values(by=0, ascending=True)

    
    select_edges = df.iloc[num_node:,:].sort_values(by=0, ascending=False).iloc[0:10,:].index.values
    for e in select_edges:
        #print("{} links the pair of nodes:".format(e))
        #pair = edge_ind[int(e[4:])-1]
        #print(nodes[pair[0]-1],nodes[pair[1]-1])
        print('Important edge links the pair of nodes:')
        print(e)
    if plot == 'all':
        plot_figure(df_sort, text='edges and nodes',fn = filename)
        return df_sort
    elif plot == 'node':
        df_node = df.iloc[0:num_node,:].sort_values(by=0, ascending=True)
        plot_figure(df_node, text='nodes', mode=mode, fn = filename)
        return df_node
    elif plot == 'edge':
        df_edge = df.iloc[num_node:,:].sort_values(by=0, ascending=True)
        plot_figure(df_edge, text='edges',mode=mode, fn = filename)
        return df_edge

def plot_figure(df,text='edges and nodes',mode='accuracy', fn = None):
    
    plt.figure(figsize=(25,15))
    plt.barh(df.iloc[-20:,:].index.values, df.iloc[-20:,:][0]*100)
    for y, (x, c) in enumerate(zip(df.iloc[-20:,:][0]*100+0.4, df.iloc[-20:,:][0]*100)):
        plt.text(x, y, str(round(c,4))+"%", ha='center', va='center')
    plt.xlabel("relative increasing {} (%)".format(mode))
    plt.title("Importance of {}".format(text))
    plt.savefig("img/vi/" + '.'.join(fn.split("/")[-1].split(".")[0:3]) +".{}.eps".format(mode),format="eps")
    plt.show()
```
